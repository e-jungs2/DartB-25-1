```python
import numpy as np

# 마스크 정의
mask = (fatigue_df['IP'] == 0) & ((fatigue_df['H'] + fatigue_df['BB']) > 0)

# 결측 인디케이터터
fatigue_df['WHIP_missing_flag'] = fatigue_df['WHIP'].isna().astype(int)

# 할당 전 NaN 개수와 마스크 True 개수
before_nan = fatigue_df['WHIP'].isna().sum()
print("전체 NaN:", before_nan)
print("mask True 개수:", mask.sum())


#최대 WHIP 값(예: max_whip*2) 구하기
max_whip = fatigue_df.loc[fatigue_df['IP'] > 0, 'WHIP'].max()

# 할당 시도
fatigue_df.loc[mask, 'WHIP'] = max_whip * 2

# 할당 후 NaN 개수
after_nan = fatigue_df['WHIP'].isna().sum()
print("할당 후 NaN:", after_nan)

# 할당된 샘플 몇 개만 출력해서 WHIP 확인
print(fatigue_df.loc[mask, ['Year','Date','IP','H','BB','WHIP']].head(5))
```

```python
# WHIP이 NaN인 행 드랍
fatigue_df = fatigue_df.dropna(subset=['WHIP']).copy()

# 확인: 결측 0개 되어야 함
print(fatigue_df['WHIP'].isna().sum())  # ➜ 0
```
